```
import sys
sys.path.append("../src")
from bayesian_tests import bayesian_ab_test_proportions
import matplotlib.pyplot as plt
import seaborn as sns
```

```
res = bayesian_ab_test_proportions(100, 1000, 120, 1000)
print(f"Prob B > A: {res[\"prob_b_better\"]:.2%}")
print(f"Expected Loss (Choose B): {res[\"expected_loss_b\"]:.4f}")

sns.kdeplot(res["samples_a"], label="Group A", fill=True)
sns.kdeplot(res["samples_b"], label="Group B", fill=True)
plt.legend()
plt.title("Posterior Distributions")
plt.show()
```
