# Phase 1: Statistical Foundations of A/B Testing

This notebook covers the fundamental concepts of A/B testing.

## 1. Hypothesis Testing
- **Null Hypothesis (H0)**: No difference between groups.
- **Alternative Hypothesis (H1)**: There is a difference.
- **Type I Error**: False positive (rejecting H0 when true). Managed by alpha (usually 0.05).
- **Type II Error**: False negative (failing to reject H0 when false). Managed by beta. Power = 1 - beta.

## 2. Common Pitfalls
- **Peeking Problem**: Checking p-values continuously inflates Type I error.
- **Multiple Testing**: Testing many variants or metrics increases false positives.
- **Sample Ratio Mismatch (SRM)**: When actual traffic split differs significantly from intended.

```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Visualize Alpha and Beta
x = np.linspace(-4, 4, 1000)
y_h0 = stats.norm.pdf(x, 0, 1)
y_h1 = stats.norm.pdf(x, 2, 1)

plt.figure(figsize=(10,6))
plt.plot(x, y_h0, label="H0 (Control)")
plt.plot(x, y_h1, label="H1 (Treatment)")
plt.axvline(1.645, color="red", linestyle="--", label="Alpha Threshold")
plt.fill_between(x, 0, y_h0, where=(x > 1.645), color="red", alpha=0.3, label="Type I Error (Alpha)")
plt.fill_between(x, 0, y_h1, where=(x < 1.645), color="blue", alpha=0.3, label="Type II Error (Beta)")
plt.legend()
plt.title("Visualizing Type I and Type II Errors")
plt.show()
```
