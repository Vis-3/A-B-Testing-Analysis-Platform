# Phase 2: Frequentist A/B Testing
Classical statistical methods.

```
import sys
sys.path.append("../src")
from frequentist_tests import z_test_proportions, t_test_continuous
from utils import generate_binary_experiment_data, generate_continuous_experiment_data
```

## Proportions (Z-Test)

```
data = generate_binary_experiment_data(1000, 1000, 0.10, 0.12)
succ_a = data[data["group"]=="A"]["converted"].sum()
tot_a = len(data[data["group"]=="A"])
succ_b = data[data["group"]=="B"]["converted"].sum()
tot_b = len(data[data["group"]=="B"])

results = z_test_proportions(succ_a, tot_a, succ_b, tot_b)
print(results)
```

## Continuous Metrics (T-Test)

```
data_c = generate_continuous_experiment_data(1000, 1000, 50, 52, 10)
val_a = data_c[data_c["group"]=="A"]["value"]
val_b = data_c[data_c["group"]=="B"]["value"]

results_c = t_test_continuous(val_a, val_b)
print(results_c)
```
